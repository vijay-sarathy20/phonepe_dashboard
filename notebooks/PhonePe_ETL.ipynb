```python
# previously we have made the json file aggregated/ transaction and user into list of dictionaries and converted into dataframe using pandas and saved as an csv file.

# Now we're going to make the json file aggregated/user into list of dictionaries and convert it into dataframe using pandas and save it as an csv file for sql usage.

'''
but we have already made aggregated_user.csv then why again because inside user data we have year wise data year wise and state wise but now don't consider state wise,
because that's an different business case. Inside the year we have quarterly file called 1.json, 2.json, 3.json, 4.json in each year. Inside that we have only extracted
aggregated portion from the user data which is registeredUsers and appOpens. Now we have to extract the userByDevice data from that to analysis the business case studies
based on phone usage and more business studies on phonePe. Here we're primarly targeting the second business case study on the project file.
'''
```

```python
# Now we're extracting the userByDevice data:

records = []

for year in os.listdir(base_path1):
  if not year.isdigit():
    continue

  year_path = os.path.join(base_path1,year)

  for file in os.listdir(year_path):
    if file.endswith('.json'):
      quarter = 'Q'+ file.replace('.json', '')

      file_path = os.path.join(year_path,file)

      with open(file_path,'r') as f:
        data = json.load(f)

      devices = data['data']['usersByDevice']

      if devices is None:
        continue

      for device in devices:
        records.append({
            'year' : int(year),
            'quarter' : quarter,
            'brand' : device['brand'],
            'count' : device['count'],
            'percentage' : device['percentage']
        })


df_user_device = pd.DataFrame(records)
df_user_device.head()


```

```python
# Data cleaning and standardization:

# Now, we're going to define a function to clean the created dataframe:

def clean_phonepe_data(df):

  # first, standardize column names:
  df.columns = df.columns.str.lower()

  # Now, clean the state names:

  if 'state' in df.columns:
    df['state'] = (
        df['state']
        .str.replace('-', '', regex = False)
        .str.strip()
        .str.title()
    )

  # Now, clean the district names:

  if 'district' in df.columns:
    df['district'] = (
        df['district']
        .str.replace(' district', '', regex = False)
        .str.strip()
        .str.title()
    )

  # Then, we have to ensure the numeric columns for sql practicing:

  numeric_columns = ['count', 'amount', 'registeredusers']

  for column in numeric_columns:
    if column in df.columns:
      df[column] = pd.to_numeric(df[column], errors = 'coerce')


  return df
```
